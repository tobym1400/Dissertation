```python
import os
import pandas as pd
import openpyxl

data_path_6_1 = "Shanghai-Telcome-Six-Months-DataSet/data_6.1~6.15.xlsx"
data_path_6_2 = "Shanghai-Telcome-Six-Months-DataSet/data_6.16~6.30.xlsx"

df_1 = pd.read_excel(data_path_6_1)
df_2 = pd.read_excel(data_path_6_2)
df = df_1.append(df_2, ignore_index=True)
```

```python
#Print percentage of nan values for each day
minVal = 10
minDay = 100

for i in range(1,31):
    percent = df[df["date"] == i].isnull().sum().sum()/df[df["date"] == i].count().sum()
    print("Day",i,":",percent)
    if percent < minVal:
        minVal = percent
        minDay = i
        
print("Lowest percentage: Day",minDay)

#Lowest percentage is day 13 - All days are fairly similar in terms of percentage
```

```python
#Print number of values per day
maxVal = 0
maxDay = 100

for i in range(1,31):
    number = df[df["date"] == i].count().sum()
    print("Day",i,":",number)
    if number > maxVal:
        maxVal = number
        maxDay = i
        
print("Highest amount: Day",maxDay)

#Highest amount is day 13 - Most days are fairly similar in terms of percentage
```

```python
import os
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import numpy as np

data_path_6_1 = "Shanghai-Telcome-Six-Months-DataSet/data_6.1~6.15.xlsx"
df = pd.read_excel(data_path_6_1)
```

```python
sample = df[df["date"] == 13]

#Duration
sample["duration"] = sample["end time"] - sample["start time"]
sample["duration_secs"] = sample["duration"].dt.total_seconds()

train = sample

#Requests per min
totalForEachMinute = np.zeros(1440)
for index,row in train.iterrows():
    minuteOfStart = row["start time"].hour * 60 + row["start time"].minute
    totalForEachMinute[minuteOfStart] += 1

#Requests per min graph
x = np.linspace(0,1440,1440)
plt.plot(x,totalForEachMinute)
plt.show()

#Remove outliers from requests per min
mean = np.mean(totalForEachMinute)
sd = np.std(totalForEachMinute)
dist_from_mean = abs(totalForEachMinute - mean)
max_deviations = 2
not_outlier = dist_from_mean < max_deviations * sd
no_outliers = totalForEachMinute[not_outlier]

#Request per min graph - no outliers
x = np.linspace(0,1440,1435)
plt.plot(x,no_outliers)
plt.show()

#Duration
print("Mean",train["duration_secs"].mean()/60)
print("Mode",train["duration_secs"].mode()/60)
print("Max",train["duration_secs"].max()/60)
```

```python
#Consider pipeline to remove outliers from dataframe
#Justify hours and test different timwe splits

#Sample Hours
decreasing = train[train["start time"].dt.hour == 1]
increasing = train[train["start time"].dt.hour == 9]
fluctuating = train[train["start time"].dt.hour == 16]

print(decreasing.info())
print(increasing.info())
print(fluctuating.info())

#Request per min - decreasing
dec_totalForEachMinute = np.zeros(60)
for index,row in decreasing.iterrows():
    minuteOfStart = row["start time"].minute
    dec_totalForEachMinute[minuteOfStart] += 1

#Request per min graph - decreasing
x = np.linspace(0,60,60)
plt.plot(x,dec_totalForEachMinute)
z = np.polyfit(x, dec_totalForEachMinute, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show()

#Request per min - increasing
inc_totalForEachMinute = np.zeros(60)
for index,row in increasing.iterrows():
    minuteOfStart = row["start time"].minute
    inc_totalForEachMinute[minuteOfStart] += 1

#Request per min graph - increasing
x = np.linspace(0,60,60)
plt.plot(x,inc_totalForEachMinute)
z = np.polyfit(x, inc_totalForEachMinute, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show()

#Request per min - fluctuating
fluc_totalForEachMinute = np.zeros(60)
for index,row in fluctuating.iterrows():
    minuteOfStart = row["start time"].minute
    fluc_totalForEachMinute[minuteOfStart] += 1
    
#Request per min graph - fluctuating
x = np.linspace(0,60,60)
plt.plot(x,fluc_totalForEachMinute)
z = np.polyfit(x,fluc_totalForEachMinute, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn import svm
from sklearn.neighbors import KNeighborsRegressor

def slidingWindow(arr, k):
    n=len(arr)
    rtnArray=[]
    for x in range(n-k):
        column_elements=[]
        for y in range(k):
            column_elements.append(arr[x+y])
        rtnArray.append(column_elements)
    return rtnArray

dec_lin_reg=LinearRegression()
inc_lin_reg=LinearRegression()
fluc_lin_reg=LinearRegression()
dec_svm=svm.SVR(kernel="rbf",C=1.0)
inc_svm=svm.SVR(kernel="rbf",C=1.0)
fluc_svm=svm.SVR(kernel="rbf",C=1.0)
dec_K = KNeighborsRegressor()
inc_K = KNeighborsRegressor()
fluc_K = KNeighborsRegressor()

dec=[]
inc=[]
fluc=[]

def results (labels, predictions):
    mae=mean_absolute_error(labels, predictions)
    mape=mean_absolute_percentage_error(labels, predictions)
    mse=mean_squared_error(labels, predictions)
    rmse=np.sqrt(mse)    
    return mae,mape,rmse
```

```python
from sklearn.model_selection import GridSearchCV

for i in range (2,10):
    window = np.array(slidingWindow(dec_totalForEachMinute,i))
    dec_totalForEachMinute_T = (np.array(dec_totalForEachMinute[i:])[np.newaxis]).T
    dec_data = np.append(window,dec_totalForEachMinute_T,axis=1)

    dec_train_set, dec_test_set = train_test_split(dec_data, test_size=0.2, random_state=42)
    dec_training = dec_train_set[:,:i]
    dec_training_labels = dec_train_set[:,i]
    dec_testing = dec_test_set[:,:i]
    dec_testing_labels = dec_test_set[:,i]

    param_grid  = [{"n_neighbors": [5,10,15,20,25],
                    "algorithm" : ['ball_tree'],
                    "leaf_size" : [10,15,20], "weights" : ['uniform']}]


    grid_search = GridSearchCV(dec_K, param_grid, cv=5,scoring='neg_mean_squared_error')
    grid_search.fit(dec_training, dec_training_labels)
    #print(str(i)+" "+str(grid_search.best_params_))
    
dec_K = KNeighborsRegressor(n_neighbors=25, algorithm='ball_tree',
                    leaf_size=20, weights='uniform')
inc_K = KNeighborsRegressor(n_neighbors=25, algorithm='ball_tree',
                    leaf_size=20, weights='uniform')
fluc_K = KNeighborsRegressor(n_neighbors=25, algorithm ='ball_tree',
                    leaf_size=20, weights='uniform')
```

```python
for i in range (2,10):
    window = np.array(slidingWindow(dec_totalForEachMinute,i))
    dec_totalForEachMinute_T = (np.array(dec_totalForEachMinute[i:])[np.newaxis]).T
    dec_data = np.append(window,dec_totalForEachMinute_T,axis=1)

    window = np.array(slidingWindow(inc_totalForEachMinute,i))
    inc_totalForEachMinute_T = (np.array(inc_totalForEachMinute[i:])[np.newaxis]).T
    inc_data = np.append(window,inc_totalForEachMinute_T,axis=1)

    window = np.array(slidingWindow(fluc_totalForEachMinute,i))
    fluc_totalForEachMinute_T = (np.array(fluc_totalForEachMinute[i:])[np.newaxis]).T
    fluc_data = np.append(window,fluc_totalForEachMinute_T,axis=1)

    #Train/test split
    dec_train_set, dec_test_set = train_test_split(dec_data, test_size=0.2, random_state=42)
    inc_train_set, inc_test_set = train_test_split(inc_data, test_size=0.2, random_state=42)
    fluc_train_set, fluc_test_set = train_test_split(fluc_data, test_size=0.2, random_state=42)

    #Decreasing Pre Processing
    dec_training = dec_train_set[:,:i]
    dec_training_labels = dec_train_set[:,i]
    dec_testing = dec_test_set[:,:i]
    dec_testing_labels = dec_test_set[:,i]

    #Increasing Pre Processing
    inc_training = inc_train_set[:,:i]
    inc_training_labels = inc_train_set[:,i]
    inc_testing = inc_test_set[:,:i]
    inc_testing_labels = inc_test_set[:,i]

    #Fluctuating Pre Processing
    fluc_training = fluc_train_set[:,:i]
    fluc_training_labels = fluc_train_set[:,i]
    fluc_testing = fluc_test_set[:,:i]
    fluc_testing_labels = fluc_test_set[:,i] 

    #Decreasing
    #LR
    dec_lin_reg.fit(dec_training, dec_training_labels)
    dec_testing_predictions = dec_lin_reg.predict(dec_testing)
    dec_LR_mae, dec_LR_mape, dec_LR_rmse = results(dec_testing_labels, dec_testing_predictions)   
    #SVM
    dec_svm.fit(dec_training, dec_training_labels)
    dec_testing_predictions = dec_svm.predict(dec_testing)
    dec_SVM_mae, dec_SVM_mape, dec_SVM_rmse = results(dec_testing_labels, dec_testing_predictions)
    #K
    dec_K.fit(dec_training, dec_training_labels)
    dec_testing_predictions = dec_K.predict(dec_testing)
    dec_K_mae, dec_K_mape, dec_K_rmse = results(dec_testing_labels, dec_testing_predictions)
    
    #Increasing
    #LR
    inc_lin_reg.fit(inc_training, inc_training_labels)
    inc_testing_predictions = inc_lin_reg.predict(inc_testing)
    inc_LR_mae, inc_LR_mape, inc_LR_rmse = results(inc_testing_labels, inc_testing_predictions)
    #SVM
    inc_svm.fit(inc_training, inc_training_labels)
    inc_testing_predictions = inc_svm.predict(inc_testing)
    inc_SVM_mae, inc_SVM_mape, inc_SVM_rmse = results(inc_testing_labels, inc_testing_predictions)
    #K
    inc_K.fit(inc_training, inc_training_labels)
    inc_testing_predictions = inc_K.predict(inc_testing)
    inc_K_mae, inc_K_mape, inc_K_rmse = results(inc_testing_labels, inc_testing_predictions)
    
    #Fluctuating
    #LR
    fluc_lin_reg.fit(fluc_training, fluc_training_labels)
    fluc_testing_predictions = fluc_lin_reg.predict(fluc_testing)
    fluc_LR_mae, fluc_LR_mape, fluc_LR_rmse = results(fluc_testing_labels, fluc_testing_predictions)
    #SVM
    fluc_svm.fit(fluc_training, fluc_training_labels)
    fluc_testing_predictions = fluc_svm.predict(fluc_testing)
    fluc_SVM_mae, fluc_SVM_mape, fluc_SVM_rmse = results(fluc_testing_labels, fluc_testing_predictions)
    #K
    fluc_K.fit(fluc_training, fluc_training_labels)
    fluc_testing_predictions = fluc_K.predict(fluc_testing)
    fluc_K_mae, fluc_K_mape, fluc_K_rmse = results(fluc_testing_labels, fluc_testing_predictions)
    
    dec.append([i,dec_LR_mae,dec_SVM_mae,dec_K_mae, dec_LR_mape, dec_SVM_mape, dec_K_mape,
                dec_LR_rmse,dec_SVM_rmse ,dec_K_rmse])
    inc.append([i,inc_LR_mae,inc_SVM_mae,inc_K_mae, inc_LR_mape, inc_SVM_mape, inc_K_mape,
                inc_LR_rmse,inc_SVM_rmse ,inc_K_rmse])
    fluc.append([i,fluc_LR_mae,fluc_SVM_mae,fluc_K_mae, fluc_LR_mape, fluc_SVM_mape, fluc_K_mape,
                 fluc_LR_rmse,fluc_SVM_rmse ,fluc_K_rmse])
```
